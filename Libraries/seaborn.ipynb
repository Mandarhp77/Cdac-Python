```python
import seaborn as sns
import numpy as np
```

```python
import matplotlib.pyplot as plt
print(sns.get_dataset_names())
```

```python
sns.set(style="darkgrid")
rs=np.random.RandomState(10)
d=rs.normal(size=100)
a=sns.displot(d,kde=True,color="b")
display(a)
```

```python
sns.set(style="ticks")
fm=sns.load_dataset("fmri")
a=sns.lineplot(x="timepoint",y="signal",hue="region",data=fm)
plt.show(a)
```

```python
sns.set(style="ticks")
fm=sns.load_dataset("fmri")
a=sns.lmplot(x="timepoint",y="signal",data=fm)
plt.show(a)
```

```python
sns.set(style="ticks")
fm=sns.load_dataset("iris")
sns.relplot(x="sepal_width",y="species",data=fm)
plt.show()
```

```python
sns.set(style="ticks")
fm=sns.load_dataset("iris")
sns.barplot(x="species",y="sepal_width",data=fm)
plt.show()
```
